```python
from scipy.io import arff
import pandas as pd

data = arff.loadarff('vote.arff')
df = pd.DataFrame(data[0])

df.head()
df.describe()
```

The data is in bytes. Let's decode it

```python
df = df.applymap(lambda x: x.decode('utf-8'))
df.head()
```
